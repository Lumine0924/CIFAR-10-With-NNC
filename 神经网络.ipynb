```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
import time
from tqdm import tqdm

class Activation:
    @staticmethod
    def relu(x):
        return np.maximum(0, x)
    
    @staticmethod
    def relu_derivative(x):
        return np.where(x > 0, 1, 0)
    
    @staticmethod
    def sigmoid(x):
        return 1 / (1 + np.exp(-np.clip(x, -500, 500)))
    
    @staticmethod
    def sigmoid_derivative(x):
        sigmoid_x = Activation.sigmoid(x)
        return sigmoid_x * (1 - sigmoid_x)
    
    @staticmethod
    def tanh(x):
        return np.tanh(x)
    
    @staticmethod
    def tanh_derivative(x):
        return 1 - np.tanh(x) ** 2
    
    @staticmethod
    def softmax(x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

class Loss:
    @staticmethod
    def cross_entropy(y_pred, y_true):
        m = y_true.shape[0]
        log_likelihood = -np.log(y_pred[range(m), y_true.argmax(axis=1)] + 1e-8)
        return np.sum(log_likelihood) / m
    
    @staticmethod
    def cross_entropy_derivative(y_pred, y_true):
        m = y_true.shape[0]
        grad = y_pred.copy()
        grad[range(m), y_true.argmax(axis=1)] -= 1
        return grad / m

class SGDOptimizer:
    def __init__(self, learning_rate=0.01, momentum=0.9, lr_decay=0.95, weight_decay=0.0001):
        self.learning_rate = learning_rate
        self.initial_learning_rate = learning_rate
        self.momentum = momentum
        self.lr_decay = lr_decay
        self.weight_decay = weight_decay
        self.velocities = {}
    
    def update(self, params, grads, iteration):
        if iteration > 0 and iteration % 100 == 0:
            self.learning_rate *= self.lr_decay
        
        if not self.velocities:
            for key in params:
                self.velocities[key] = np.zeros_like(params[key])
        
        for key in params:
            l2_grad = self.weight_decay * params[key]
            self.velocities[key] = self.momentum * self.velocities[key] - self.learning_rate * (grads[key] + l2_grad)
            params[key] += self.velocities[key]
        
        return params

class ThreeLayerNN:
    def __init__(self, input_size, hidden_size, output_size, activation='relu'):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        self.params = {
            'W1': np.random.randn(input_size, hidden_size) * np.sqrt(2. / input_size),
            'b1': np.zeros((1, hidden_size)),
            'W2': np.random.randn(hidden_size, output_size) * np.sqrt(2. / hidden_size),
            'b2': np.zeros((1, output_size))
        }
        
        if activation == 'relu':
            self.activation = Activation.relu
            self.activation_derivative = Activation.relu_derivative
        elif activation == 'sigmoid':
            self.activation = Activation.sigmoid
            self.activation_derivative = Activation.sigmoid_derivative
        elif activation == 'tanh':
            self.activation = Activation.tanh
            self.activation_derivative = Activation.tanh_derivative
        else:
            raise ValueError(f"Unsupported activation function: {activation}")
        
        self.cache = {}
    
    def forward(self, X):
        Z1 = np.dot(X, self.params['W1']) + self.params['b1']
        A1 = self.activation(Z1)
        
        Z2 = np.dot(A1, self.params['W2']) + self.params['b2']
        A2 = Activation.softmax(Z2)
        
        self.cache = {
            'X': X,
            'Z1': Z1,
            'A1': A1,
            'Z2': Z2,
            'A2': A2
        }
        
        return A2
    
    def backward(self, y):
        m = y.shape[0]
        
        dZ2 = self.cache['A2'] - y
        dW2 = np.dot(self.cache['A1'].T, dZ2) / m
        db2 = np.sum(dZ2, axis=0, keepdims=True) / m
        
        dA1 = np.dot(dZ2, self.params['W2'].T)
        dZ1 = dA1 * self.activation_derivative(self.cache['Z1'])
        dW1 = np.dot(self.cache['X'].T, dZ1) / m
        db1 = np.sum(dZ1, axis=0, keepdims=True) / m
        
        grads = {
            'W1': dW1,
            'b1': db1,
            'W2': dW2,
            'b2': db2
        }
        
        return grads
    
    def predict(self, X):
        A2 = self.forward(X)
        return np.argmax(A2, axis=1)
    
    def save_model(self, filepath):
        with open(filepath, 'wb') as f:
            pickle.dump(self.params, f)
    
    def load_model(self, filepath):
        with open(filepath, 'rb') as f:
            self.params = pickle.load(f)

def load_cifar10():
    def unpickle(file):
        with open(file, 'rb') as fo:
            dict = pickle.load(fo, encoding='bytes')
        return dict
    
    if not os.path.exists('./cifar-10-batches-py'):
        print("CIFAR-10数据集不存在，正在自动下载...")
        
        import urllib.request
        import tarfile
        
        url = "https://www.cs.toronto.edu/~kriz/cifar-10-python.tar.gz"
        filename = "cifar-10-python.tar.gz"
        
        try:
            def report_progress(block_num, block_size, total_size):
                downloaded = block_num * block_size
                percent = min(100, downloaded * 100 / total_size)
                print(f"\r下载进度: {percent:.1f}% ({downloaded}/{total_size} 字节)", end="")
            
            urllib.request.urlretrieve(url, filename, reporthook=report_progress)
            print("\n下载完成！正在解压...")
            
            with tarfile.open(filename, 'r:gz') as tar:
                tar.extractall()
            print("解压完成！")
            
            os.remove(filename)
            print(f"已删除压缩包 {filename}")
            
        except Exception as e:
            print(f"下载或解压过程中出错: {e}")
            print("请尝试手动下载数据集:")
            print("下载链接: https://www.cs.toronto.edu/~kriz/cifar-10-python.tar.gz")
            raise FileNotFoundError("CIFAR-10数据集未找到")
    
    X_train = []
    y_train = []
    for i in range(1, 6):
        batch = unpickle(f'./cifar-10-batches-py/data_batch_{i}')
        X_train.append(batch[b'data'])
        y_train.extend(batch[b'labels'])
    
    X_train = np.vstack(X_train)
    y_train = np.array(y_train)
    
    test_batch = unpickle('./cifar-10-batches-py/test_batch')
    X_test = test_batch[b'data']
    y_test = np.array(test_batch[b'labels'])
    
    X_train = X_train.reshape(X_train.shape[0], 3, 32, 32).transpose(0, 2, 3, 1)
    X_test = X_test.reshape(X_test.shape[0], 3, 32, 32).transpose(0, 2, 3, 1)
    
    X_train = X_train.astype('float32') / 255.0
    X_test = X_test.astype('float32') / 255.0
    
    mean = np.mean(X_train, axis=0)
    std = np.std(X_train, axis=0) + 1e-7
    X_train = (X_train - mean) / std
    X_test = (X_test - mean) / std
    
    X_train = X_train.reshape(X_train.shape[0], -1)
    X_test = X_test.reshape(X_test.shape[0], -1)
    
    num_training = 45000
    num_validation = 5000
    
    mask = list(range(num_training))
    X_val = X_train[num_training:num_training + num_validation]
    y_val = y_train[num_training:num_training + num_validation]
    X_train = X_train[mask]
    y_train = y_train[mask]
    
    def to_one_hot(y, num_classes=10):
        return np.eye(num_classes)[y]
    
    y_train = to_one_hot(y_train)
    y_val = to_one_hot(y_val)
    y_test = to_one_hot(y_test)
    
    return X_train, y_train, X_val, y_val, X_test, y_test

def train_model(model, X_train, y_train, X_val, y_val, batch_size=64, num_epochs=10, 
               optimizer_config={'learning_rate': 0.01, 'momentum': 0.9, 'lr_decay': 0.95, 'weight_decay': 0.0001}):
    
    optimizer = SGDOptimizer(**optimizer_config)
    
    train_loss_history = []
    val_loss_history = []
    val_acc_history = []
    best_val_acc = 0
    
    num_samples = X_train.shape[0]
    iterations_per_epoch = max(num_samples // batch_size, 1)
    num_iterations = num_epochs * iterations_per_epoch
    
    start_time = time.time()
    
    print(f"开始训练: {num_epochs} 轮, 每轮 {iterations_per_epoch} 批次, 批次大小 {batch_size}")
    for epoch in range(num_epochs):
        epoch_train_loss = 0
        
        shuffle_indices = np.random.permutation(num_samples)
        X_train_shuffled = X_train[shuffle_indices]
        y_train_shuffled = y_train[shuffle_indices]
        
        print(f"轮次 {epoch+1}/{num_epochs}:")
        for i in tqdm(range(iterations_per_epoch)):
            start_idx = i * batch_size
            end_idx = min((i + 1) * batch_size, num_samples)
            X_batch = X_train_shuffled[start_idx:end_idx]
            y_batch = y_train_shuffled[start_idx:end_idx]
            
            y_pred = model.forward(X_batch)
            batch_loss = Loss.cross_entropy(y_pred, y_batch)
            epoch_train_loss += batch_loss
            
            grads = model.backward(y_batch)
            
            model.params = optimizer.update(model.params, grads, epoch * iterations_per_epoch + i)
        
        epoch_train_loss /= iterations_per_epoch
        train_loss_history.append(epoch_train_loss)
        
        val_pred = model.forward(X_val)
        val_loss = Loss.cross_entropy(val_pred, y_val)
        val_loss_history.append(val_loss)
        
        val_acc = np.mean(np.argmax(val_pred, axis=1) == np.argmax(y_val, axis=1))
        val_acc_history.append(val_acc)
        
        elapsed = time.time() - start_time
        print(f"轮次 {epoch+1}/{num_epochs} 完成: 训练损失 {epoch_train_loss:.4f}, 验证损失 {val_loss:.4f}, 验证准确率 {val_acc:.4f}, 用时 {int(elapsed//60)}分 {int(elapsed%60)}秒")
        
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            model.save_model('best_model.pkl')
            print(f"模型已保存: 验证准确率提高到 {val_acc:.4f}")
    
    total_time = time.time() - start_time
    print(f"训练完成，总用时: {int(total_time//60)}分 {int(total_time%60)}秒")
    
    model.load_model('best_model.pkl')
    
    return train_loss_history, val_loss_history, val_acc_history

def test_model(model, X_test, y_test):
    test_pred = model.forward(X_test)
    test_acc = np.mean(np.argmax(test_pred, axis=1) == np.argmax(y_test, axis=1))
    print(f"测试集准确率: {test_acc:.4f}")
    return test_acc

def hyperparameter_search(X_train, y_train, X_val, y_val, input_size):
    results = []
    
    hidden_sizes = [128, 256, 512]
    learning_rates = [0.01, 0.005, 0.001]
    weight_decays = [0.0001, 0.0005, 0.001]
    activations = ['relu', 'sigmoid', 'tanh']
    
    best_val_acc = 0
    best_params = {}
    
    for hidden_size in hidden_sizes:
        for lr in learning_rates:
            for wd in weight_decays:
                for activation in activations:
                    print(f"\n测试超参数: hidden_size={hidden_size}, learning_rate={lr}, weight_decay={wd}, activation={activation}")
                    
                    model = ThreeLayerNN(input_size, hidden_size, 10, activation)
                    
                    optimizer_config = {
                        'learning_rate': lr,
                        'momentum': 0.9,
                        'lr_decay': 0.95,
                        'weight_decay': wd
                    }
                    
                    _, _, val_acc_history = train_model(
                        model, X_train, y_train, X_val, y_val,
                        batch_size=128, num_epochs=3,
                        optimizer_config=optimizer_config
                    )
                    
                    val_acc = val_acc_history[-1]
                    results.append({
                        'hidden_size': hidden_size,
                        'learning_rate': lr,
                        'weight_decay': wd,
                        'activation': activation,
                        'val_acc': val_acc
                    })
                    
                    if val_acc > best_val_acc:
                        best_val_acc = val_acc
                        best_params = {
                            'hidden_size': hidden_size,
                            'learning_rate': lr,
                            'weight_decay': wd,
                            'activation': activation
                        }
    
    print("\n超参数搜索结果:")
    for res in results:
        print(f"hidden_size={res['hidden_size']}, lr={res['learning_rate']}, wd={res['weight_decay']}, activation={res['activation']}: val_acc={res['val_acc']:.4f}")
    
    print(f"\n最佳超参数: {best_params}, 验证准确率: {best_val_acc:.4f}")
    return best_params, results

def visualize_training(train_loss_history, val_loss_history, val_acc_history):
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(train_loss_history, label='Training Loss')
    plt.plot(val_loss_history, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Loss Curves')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(val_acc_history, label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Validation Accuracy Curve')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('training_curves.png')
    plt.show()

def visualize_weights(model, save_path='weights_visualization.png'):
    W1 = model.params['W1']
    
    W1_reshaped = W1.reshape(-1, 32, 32, 3)
    
    num_filters = min(16, W1_reshaped.shape[0])
    fig, axes = plt.subplots(4, 4, figsize=(10, 10))
    
    for i in range(num_filters):
        ax = axes[i // 4, i % 4]
        weight_img = W1_reshaped[i]
        weight_img = (weight_img - weight_img.min()) / (weight_img.max() - weight_img.min())
        ax.imshow(weight_img)
        ax.set_title(f'Filter {i+1}')
        ax.axis('off')
    
    plt.tight_layout()
    plt.savefig(save_path)
    plt.show()
    
    print(f"权重可视化已保存至 {save_path}")

def main():
    print("加载CIFAR-10数据集...")
    X_train, y_train, X_val, y_val, X_test, y_test = load_cifar10()
    input_size = X_train.shape[1]
    
    search_params = input("是否进行超参数搜索？(y/n): ")
    if search_params.lower() == 'y':
        print("开始超参数搜索...")
        best_params, _ = hyperparameter_search(
            X_train[:5000],
            y_train[:5000], 
            X_val, 
            y_val, 
            input_size
        )
        
        model = ThreeLayerNN(
            input_size, 
            best_params['hidden_size'], 
            10,
            best_params['activation']
        )
        
        optimizer_config = {
            'learning_rate': best_params['learning_rate'],
            'momentum': 0.9,
            'lr_decay': 0.95,
            'weight_decay': best_params['weight_decay']
        }
    else:
        print("使用优化的默认超参数...")
        model = ThreeLayerNN(input_size, 512, 10, 'relu')
        optimizer_config = {
            'learning_rate': 0.001,
            'momentum': 0.9,
            'lr_decay': 0.98,
            'weight_decay': 0.0005
        }
    
    print("开始训练模型...")
    try:
        train_loss_history, val_loss_history, val_acc_history = train_model(
            model, X_train, y_train, X_val, y_val,
            batch_size=128, num_epochs=10,
            optimizer_config=optimizer_config
        )
    except KeyboardInterrupt:
        print("\n训练被用户中断！将使用当前的最佳模型进行评估。")
        if os.path.exists('best_model.pkl'):
            model.load_model('best_model.pkl')
            print("已加载最佳模型。")
    
    print("在测试集上评估模型...")
    test_acc = test_model(model, X_test, y_test)
    
    print("可视化训练过程...")
    visualize_training(train_loss_history, val_loss_history, val_acc_history)
    
    print("可视化模型权重...")
    visualize_weights(model)
    
    print(f"实验完成！最终测试准确率: {test_acc:.4f}")
    print("训练曲线和权重可视化已保存。")

if __name__ == "__main__":
    main()
```
